```python
def check_input(element_massiva):
    if isinstance(element_massiva, str):
        return True
    else:
        return False

def get_substring(s):
    if not s:
        return ""
    longest = ""
    n = len(s)

    # Перебираем все возможные начала подстрок
    for i in range(n):
        seen = set()  # Множество для отслеживания уникальных символов
        current = ""  # Текущая подстрока без повторений

        # Начинаем с позиции i и идем до конца строки
        for j in range(i, n):
            # Если символ уже встречался, прерываем цикл
            if s[j] in seen:
                break
            # Иначе добавляем его в множество и в текущую подстроку
            seen.add(s[j])
            current += s[j]

            # Если текущая подстрока длиннее сохраненной, обновляем
            if len(current) > len(longest):
                longest = current

    return longest

def main(arr):
    # Ищем самую длинную строку в списке
    max_len = 0
    idx = -1
    target_str = ""

    # Проходим по всем элементам списка
    for i in range(len(arr)):
        if check_input(arr[i]):
            if len(arr[i]) > max_len:
                max_len = len(arr[i])
                idx = i
                target_str = arr[i]

    if idx == -1:
        print("-1, ''")
        return

    # Находим самую длинную подстроку без повторений
    result_substr = get_substring(target_str)

    # Выводим результат
    print(f"{idx}, '{result_substr}'")

    return idx, result_substr
```
